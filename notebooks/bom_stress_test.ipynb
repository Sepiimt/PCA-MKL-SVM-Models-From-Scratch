# --- Beta Optimization Methods (BOM) Stress Test ---

```python
# --- General Imports ---
import numpy as np
import pandas as pd
import sys
from pathlib import Path
# --- Resolving Import Issue ---
notebook_dir = Path.cwd()
src_path = (notebook_dir / ".." / "src").resolve()
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
# --- Importing API Files ---
from mklsvm.svm import SVM
```

## > MAGIC Gamma Telescope Dataset

```python
#> --- Loading Processed Data --- <#
pca_telescope_x_train = np.load(r"..\\data\\splits\\telescope\\pca_telescope_x_train.npy", allow_pickle=True)
pca_telescope_x_test = np.load(r"..\\data\\splits\\telescope\\pca_telescope_x_test.npy", allow_pickle=True)
telescope_y_train = np.load(r"..\\data\\splits\\telescope\\telescope_y_train.npy")
```

### > BOM: "SimpleMKL"

```python
# --- Model Instance & BO Method Selection ---
telescope_SVM_model = SVM()
telescope_SVM_model.set_beta_optimizer("SimpleMKL")
```

```python
# --- Model Training ---
telescope_SVM_model.fit(pca_telescope_x_train, telescope_y_train,
                        max_bo_iter = 50,
                        bo_tolerance = 1e-7,
                        smo_c = 1,
                        max_smo_iter = 10**5,
                        smo_tolerance = 1e-3,
                        detailed_info=True)
```

```python
# --- Predicting ---
simplemkl_svm_prediction = telescope_SVM_model.predict(pca_telescope_x_test)
# --- Saving ---
telescope_SVM_model.save_model("..\\artifacts\\svm\\telescope\\simplemkl\\")
np.save(r"..\\artifacts\\predictions\\telescope\\simplemkl_svm_prediction", simplemkl_svm_prediction)
# --- Cleanup ---
del telescope_SVM_model, simplemkl_svm_prediction
```
