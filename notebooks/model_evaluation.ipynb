# --- Model Evaluation ---

```python
# --- General Imports ---
import pandas as pd
import numpy as np
import sys
from pathlib import Path
# --- Resolving Import Issue ---
notebook_dir = Path.cwd()
src_pca_path = (notebook_dir / ".." / "src").resolve()
if str(src_pca_path) not in sys.path:
    sys.path.insert(0, str(src_pca_path))
# --- Importing API Files ---
from mklsvm.svm import SVM
from mklsvm.metrics import *
```

## > Cancer Dataset 

```python
# --- Loading Model Data ---
cancer_SVM_model = SVM()
cancer_SVM_model.load_model(r"..\\artifacts\\svm\\cancer\\")
cancer_svm_prediction = np.load(r"..\\artifacts\\predictions\\cancer_svm_prediction.npy")
# --- Loading Train/Test Data ---
cancer_y_train = np.load(r"..\\data\\splits\\cancer\\cancer_y_train.npy")
cancer_y_test = np.load(r"..\\data\\splits\\cancer\\cancer_y_test.npy")
pca_cancer_x_test = np.load(r"..\\data\\splits\\cancer\\pca_cancer_x_test.npy", allow_pickle=True)
pca_cancer_x_train = np.load(r"..\\data\\splits\\cancer\\pca_cancer_x_train.npy", allow_pickle=True)
# --- Save Path ---
path = r"..\\results\benchmarks\\cancer\\"
```

```python
# --- MKL-SVM Evaluation ---
mkl_svm_evaluation(cancer_SVM_model, pca_cancer_x_train, cancer_y_train, cancer_y_test, cancer_svm_prediction, save=True, save_path=path)
```

```python
# --- SK-Learn SMO Evaluation ---
sklearn_semi_mkl_evaluation(cancer_SVM_model, 
                            pca_cancer_x_train, cancer_y_train, pca_cancer_x_test, cancer_y_test, 
                            beta=cancer_SVM_model.MKL_instance.beta_array, 
                            save=True, save_path=path)
```

```python
# --- Cleanuo ---
del cancer_SVM_model, cancer_svm_prediction
del cancer_y_train, cancer_y_test, pca_cancer_x_test, pca_cancer_x_train
```
