# --- PCA Model Training ---

```python
# --- General Imports ---
import numpy as np
import pandas as pd
import sys
from pathlib import Path
# --- Resolving Import Issue ---
notebook_dir = Path.cwd()
src_path = (notebook_dir / ".." / "src").resolve()
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
# --- Importing API Files ---
from mklsvm.pca import PCA
from mklsvm.utils import train_test_split
```

## > Breast Cancer Dataset

```python
cancer_y_array = np.load(r"..\\data\\processed\\cancer\\processed_cancer_y_array.npy")
cancer_x_array = np.load(r"..\\data\\processed\\cancer\\processed_cancer_x_array.npy", allow_pickle=True)
```

```python
# --- Preparing Values ---
cancer_x_train, cancer_x_test, cancer_y_train, cancer_y_test = train_test_split(cancer_x_array, cancer_y_array, shuffle=True, random_state=42)
# --- Preparing General PCA Report ---
cancer_PCA_model = PCA()
# --- Scaling X Values ---
scaled_cancer_x_train = cancer_PCA_model.scaler(cancer_x_train, fit=True)
scaled_cancer_x_test  = cancer_PCA_model.scaler(cancer_x_test, fit=False)  #> reuses train's means/stds
# --- Fitting Model ---
cancer_pca_report = cancer_PCA_model.fit(scaled_cancer_x_train, reset_model_values=True)
# --- Representing Report ---
cancer_pca_report
```

```python
# --- PCA Transformation ---
pca_cancer_x_train = cancer_PCA_model.transform(scaled_cancer_x_train, 10)
pca_cancer_x_test  = cancer_PCA_model.transform(scaled_cancer_x_test, 10)  #> same eigenvectors, no refit
```

```python
# --- Saving Processed Data
np.save(r"..\\data\\splits\\cancer\\pca_cancer_x_train", pca_cancer_x_train.to_numpy())
np.save(r"..\\data\\splits\\cancer\\pca_cancer_x_test", pca_cancer_x_test.to_numpy())
np.save(r"..\\data\\splits\\cancer\\cancer_y_train", cancer_y_train)
np.save(r"..\\data\\splits\\cancer\\cancer_y_test", cancer_y_test)
# --- Saving PCA Models ---
cancer_PCA_model.save_info(r"..\\artifacts\\pca\\cancer\\cancer_model_data")
```

```python
# --- Cleaning ---
del cancer_PCA_model
del cancer_x_train, cancer_x_test, cancer_y_train, cancer_y_test
del cancer_x_array, cancer_y_array
del scaled_cancer_x_train, scaled_cancer_x_test
del cancer_pca_report
```
