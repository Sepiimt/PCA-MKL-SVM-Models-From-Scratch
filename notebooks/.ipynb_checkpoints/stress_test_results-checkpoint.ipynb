# --- Stress Test Results ---

```python
# --- General Imports ---
import pandas as pd
import numpy as np
import sys
from pathlib import Path
# --- Resolving Import Issue ---
notebook_dir = Path.cwd()
src_pca_path = (notebook_dir / ".." / "src").resolve()
if str(src_pca_path) not in sys.path:
    sys.path.insert(0, str(src_pca_path))
# --- Importing API Files ---
from mklsvm.svm import SVM
from mklsvm.metrics import mkl_svm_evaluation
```

## > MAGIC Gamma Telescope Dataset

```python
# --- Loading Train/Test Data ---
telescope_y_train = np.load(r"..\\data\\splits\\telescope\\telescope_y_train.npy")
telescope_y_test = np.load(r"..\\data\\splits\\telescope\\telescope_y_test.npy")
pca_telescope_x_test = np.load(r"..\\data\\splits\\telescope\\pca_telescope_x_test.npy", allow_pickle=True)
pca_telescope_x_train = np.load(r"..\\data\\splits\\telescope\\pca_telescope_x_train.npy", allow_pickle=True)
```

### > BOM: "SimpleMKL"

```python
telescope_SVM_model = SVM()
telescope_SVM_model.load_model(r"..\\artifacts\\svm\\telescope\\simplemkl\\")
simplemkl_svm_prediction = np.load(r"..\\artifacts\\predictions\\telescope\\simplemkl_svm_prediction.npy")

path = r"..\\results\benchmarks\\telescope\\simplemkl\\"
```

```python
mkl_svm_evaluation(telescope_SVM_model, pca_telescope_x_train, telescope_y_train, telescope_y_test, simplemkl_svm_prediction, save=True, save_path=path)
```

```python
telescope_SVM_model.MKL_instance.SMO_model.b
```

```python
del telescope_SVM_model, simplemkl_svm_prediction
```
