```python
# --- General Imports ---
import pandas as pd
import numpy as np
import sys
from pathlib import Path
# --- Resolving Import Issue ---
notebook_dir = Path.cwd()
src_path = (notebook_dir / ".." / "src").resolve()
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
# --- Importing API Files ---
from mklsvm.pca import PCA
from mklsvm.utils import train_test_split
```

```python
# --- Reading and Repersenting the Dataset --- #
data_frame = pd.read_csv(r'..\data\raw\cancer_data.csv')
index_list=[i for i in range(1,len(data_frame)+1)]
data_frame.index=index_list
raw_data_amount=len(data_frame)
# --- Representing Data ---
data_frame.iloc[:,] #> Showing some part of te dataframe
```

```python
# --- Changing Values ---
diagnosis_array = data_frame["diagnosis"].to_numpy()
diagnosis_array = np.where(diagnosis_array=='M', -1, +1)
# --- Deleting Irrelevent Column ---
del data_frame["id"]
del data_frame["diagnosis"]
del data_frame["Unnamed: 32"]
# --- Representing Data ---
data_frame.iloc[:,] #> Showing some part of te dataframe
```

```python
# --- Preparing Values ---
X = data_frame.to_numpy()
data_frame_columns = data_frame.columns
x_train, x_test, y_train, y_test = train_test_split(X, diagnosis_array)
```

```python
# --- Preparing General PCA Report ---
general_PCA = PCA()
scaled_X = general_PCA.scaler(X)
PCA_X_report = general_PCA.fit(scaled_X, reset_model_values=True)
del general_PCA, scaled_X
PCA_X_report
```

```python
# --- Creating Instances ---
train_split_PCA = PCA()
test_split_PCA = PCA()
# --- Scaling X Values ---
scaled_x_train = train_split_PCA.scaler(x_train)
scaled_x_test = test_split_PCA.scaler(x_test)
# --- Fitting Models ---
_ = train_split_PCA.fit(scaled_x_train, orig_features = data_frame_columns, reset_model_values=True)
_ = test_split_PCA.fit(scaled_x_test, orig_features = data_frame_columns, reset_model_values=True)
# --- PCA Transformation ---
x_train_pca_dataframe = train_split_PCA.transform(scaled_x_train, 10)
x_test_pca_dataframe = test_split_PCA.transform(scaled_x_test, 10)
```

```python
# --- Saving Processed Data
np.save(r"..\\data\\splits\\x_train", x_train_pca_dataframe.to_numpy())
np.save(r"..\\data\\splits\\x_test", x_test_pca_dataframe.to_numpy())
np.save(r"..\\data\\splits\\y_train", y_train)
np.save(r"..\\data\\splits\\y_test", y_test)
```

```python
# --- Saving PCA Models ---
train_split_PCA.save_info(r"..\\artifacts\\pca\\train_pca_model_data")
test_split_PCA.save_info(r"..\\artifacts\\pca\\test_pca_model_data")
```
