# --- Data Cleaning & Engineering ---

```python
# --- General Imports ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

## > Breast Cancer Dataset

```python
# --- Reading and Repersenting the Dataset --- #
cancer_dataset_df = pd.read_csv(r'..\\data\\raw\\breast_cancer_dataset.csv')
# --- Checking for NaN Values ---
if len(cancer_dataset_df) == len(cancer_dataset_df.dropna()):
    print("The Dataset Has No NaN Values")
else:
    print("Dataset Needs Cleaning")
# --- Representing Data ---
cancer_dataset_df.iloc[:,] #> Showing some part of te dataframe
```

```python
# --- Changing Values ---
diagnosis_array = cancer_dataset_df["diagnosis"].to_numpy().astype(str)
cancer_results_y_array = np.where(diagnosis_array=='B', +1, -1)
# --- Cleaning Up ---
del cancer_dataset_df["id"], cancer_dataset_df["Unnamed: 32"], cancer_dataset_df["diagnosis"]
del diagnosis_array
# --- Saving ---
np.save(r"..\\data\\processed\\cancer\\processed_cancer_y_array", cancer_results_y_array)
np.save(r"..\\data\\processed\\cancer\\processed_cancer_x_array", cancer_dataset_df.to_numpy())
# --- Cleaning Up ---
del cancer_results_y_array, cancer_dataset_df
```

## > MAGIC Gamma Telescope Dataset

```python
# --- Reading and Repersenting the Dataset --- #
telescope_dataset_df = pd.read_csv(r'..\\data\\raw\\magic_gamma_telescope.csv')
# --- Checking for NaN Values ---
if len(telescope_dataset_df) == len(telescope_dataset_df.dropna()):
    print("The Dataset Has No NaN Values")
# --- Representing Data ---
telescope_dataset_df.iloc[:,] #> Showing some part of te dataframe
```

```python
# --- Changing Values ---
telescope_results_array = telescope_dataset_df["class"].to_numpy().astype(str)
telescope_results_array = np.where(telescope_results_array=='g', +1, -1)
# --- Cleaning Up ---
del telescope_dataset_df["class"]
# --- Reducing Dataset's length To Prevent Memory Overload ---
processed_telescope_x_array, _, processed_telescope_y_array, _ = train_test_split(telescope_dataset_df.to_numpy(), telescope_results_array, 
                                                                                  train_size=4000, stratify=telescope_results_array, random_state=42)
# --- Saving ---
np.save(r"..\\data\\processed\\telescope\\processed_telescope_y_array", processed_telescope_y_array)
np.save(r"..\\data\\processed\\telescope\\processed_telescope_x_array", processed_telescope_x_array)
# --- Cleaning Up ---
del telescope_results_array, telescope_dataset_df, processed_telescope_y_array, processed_telescope_x_array
```
