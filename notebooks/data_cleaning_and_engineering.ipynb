```python
# --- General Imports ---
import pandas as pd
import numpy as np
import sys
from pathlib import Path
# --- Resolving Import Issue ---
notebook_dir = Path.cwd()
src_path = (notebook_dir / ".." / "src" / "pca").resolve()
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
# --- Importing API Files ---
from pca import PCA
```

```python
# --- Reading and Repersenting the Dataset --- #
data_frame = pd.read_csv(r'..\data\raw\cancer_data.csv')
index_list=[i for i in range(1,len(data_frame)+1)]
data_frame.index=index_list
raw_data_amount=len(data_frame)
# --- Representing Data ---
data_frame.iloc[:,] #> Showing some part of te dataframe
```

```python
# --- Changing Values ---
diagnosis_array = data_frame["diagnosis"].to_numpy()
diagnosis_array = np.where(diagnosis_array=='M', -1, +1)
# --- Deleting Irrelevent Column ---
del data_frame["id"]
del data_frame["diagnosis"]
del data_frame["Unnamed: 32"]
# --- Representing Data ---
data_frame.iloc[:,] #> Showing some part of te dataframe
```

```python
# --- Preparing PCA Parameters ---
PCA_model = PCA()
X = data_frame.to_numpy()
data_frame_columns = data_frame.columns
scaled_X = PCA_model.scaler(X)
```

```python
# --- PCA Report ---
PCA_X_report = PCA_model.fit(scaled_X, orig_features = data_frame_columns, reset_model_values=True)
PCA_X_report
```

```python
# --- Initiating PCA ---
PCA_data_frame = PCA_model.transform(scaled_X, 10)
PCA_data_frame
```

```python
# --- Saving Processed Data ---
PCA_data_frame.to_csv(r'..\\data\\processed\\processed_and_PCA_transformed_X.csv', index=False)
np.save(r'..\\data\\processed\\processed_and_SVM_encoded_Y', diagnosis_array)
PCA_model.save_info(r"..\\data\\pca\\pca_data")
```
