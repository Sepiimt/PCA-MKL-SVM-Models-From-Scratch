```python
# --- General Imports ---
import numpy as np
import pandas as pd
import csv
import sys
from pathlib import Path
# --- Resolving Import Issue ---
notebook_dir = Path.cwd()
src_path = (notebook_dir / ".." / "src" / "svm").resolve()
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
# --- Importing API Files ---
from svm import SVM
from utils import Train_Test_Split
from metrics import MKL_SVM_Evaluation
```

```python
#> --- Loading Processed Data --- <#
PCA_transformed_X_df = pd.read_csv(r'..\\data\\processed\\processed_and_PCA_transformed_X.csv')
SVM_encoded_Y = np.load(r'..\\data\\processed\\processed_and_SVM_encoded_Y.npy')
```

```python
PCA_transformed_X_df
```

```python
PCA_transformed_X_array = PCA_transformed_X_df.to_numpy()
x_train, x_test, y_train, y_test = Train_Test_Split(PCA_transformed_X_array, SVM_encoded_Y)
```

```python
SVM_model = SVM()
SVM_model.select_kernels([True, True, True, True])
SVM_model.set_kernels_arguments(x_train)
```

```python
SVM_model.fit(x_train, y_train,
              beta_learning_rate = 0.1, max_beta_optimization_iter = 1200, beta_optimization_tolerance = 1e-3,
              smo_C = 3, max_SMO_iter = 800, SMO_tolerance = 1e-3,
              general_timer=True, detailed_timer=False)
```

```python
model_prediction_result = SVM_model.predict(x_test)
SVM_model.save_model("..\\data\\svm\\")
np.save(r"..\\data\\seperated\\x_train", x_train)
np.save(r"..\\data\\seperated\\x_test", x_test)
np.save(r"..\\data\\seperated\\y_train", y_train)
np.save(r"..\\data\\seperated\\y_test", y_test)
np.save(r"..\\data\\svm\\y_predicted", model_prediction_result)
```
